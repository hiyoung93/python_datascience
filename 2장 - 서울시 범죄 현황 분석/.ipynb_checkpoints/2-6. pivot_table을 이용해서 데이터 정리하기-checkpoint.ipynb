```python
import pandas as pd
import numpy as np
```

```python
crime_anal_raw = pd.read_csv('02. crime_in_Seoul_include_gu_name.csv',
                            encoding='utf-8')
crime_anal_raw.head()
```

```python
crime_anal_raw = pd.read_csv('02. crime_in_Seoul_include_gu_name.csv',
                            encoding='utf-8', index_col = 0)
crime_anal = pd.pivot_table(crime_anal_raw, index = '구별', aggfunc=np.sum)
# row 의 index를 '구별'로 가져가고, aggfunc로 구별끼리 합한다.
crime_anal.head()
```

```python
crime_anal['강간검거율'] = crime_anal['강간 검거']/crime_anal['강간 발생'] *100
crime_anal['강도검거율'] = crime_anal['강도 검거']/crime_anal['강도 발생'] *100
crime_anal['살인검거율'] = crime_anal['살인 검거']/crime_anal['살인 발생'] *100
crime_anal['절도검거율'] = crime_anal['절도 검거']/crime_anal['절도 발생'] *100
crime_anal['폭력검거율'] = crime_anal['폭력 검거']/crime_anal['폭력 발생'] *100


del crime_anal['강간 검거']
del crime_anal['강도 검거']
del crime_anal['살인 검거']
del crime_anal['절도 검거']
del crime_anal['폭력 검거']
```

검거**율** 계산<br>
검거**건수** 는 검거**율**로 대체 가능하니까 **삭제**하기

```python
crime_anal.head()
```

```python
con_list = ['강간검거율','강도검거율','살인검거율','절도검거율','폭력검거율']

for column in con_list:
    crime_anal.loc[crime_anal[column] > 100, column] = 100
    
crime_anal.head()
```

```python
crime_anal.rename(columns = {'강간 발생':'강간',
                            '강도 발생':'강도',
                            '살인 발생':'살인',
                            '절도 발생':'절도',
                            '폭력 발생':'폭력'}, inplace = True)
```

```python
crime_anal.head()
```

## scikit-Learn 
### preprocessing(전처리 담당하는 서브패키지)
- StandardScaler(X) : 평균이 0 과 표준편자가 1이 되도록 변환
- RobustScaler(X) : 중앙값이 0, IQR(interquartile range,사분범위) 1
- MinMaxScaler(X) : 최대값이 각각1, 죄소값이 0이 되도록 변환
- MaxAbsScaler(X) : 0을 기준으로 절대값이 가장 큰 수가 1 또는 -1이 되도록 변환

1. 학습용 데이터 분포 추정 : 학습용데이터 입력 fit메서드 실행 분포 모수를 객체내에 저장

2. 학습용 데이터 변환 : 학습용 데이터 입력, transform실행하고 학습용 데이터 변환

3. 검증용 데이터 변환 : 검증용 데이터 입력해서 transform실행하면 검증용 데이터 변환


1 + 2 = fit_transform사용가능

```python
from sklearn import preprocessing
```

```python
import warnings
warnings.filterwarnings('ignore')
# warnings massege안나오게 함
```

```python
col = ['강간','강도','살인','절도','폭력']

x = crime_anal[col].values
min_max_scaler = preprocessing.MinMaxScaler()
# preprocessing.MinMaxScaler() = 
    # preprocessing - 전처리 담장 scikit-learn 서브 패키지
    # MinMaxScaler() - 최대값이 1 최소값이 0 이 되도록 변환
```

```python
x_scaled = min_max_scaler.fit_transform(x.astype(float))
# preprocessing.MinMacScaler()를 저장한 변수에다가
# 변수.fit_transform
crime_anal_norm = pd.DataFrame(x_scaled, 
                              columns = col,
                              index = crime_anal.index)
```

```python
col2 = ['강간검거율','강도검거율','살인검거율','절도검거율','폭력검거율']
crime_anal_norm[col2] = crime_anal[col2]
crime_anal_norm.head()
```

```python
result_CCTV = pd.read_csv('01. CCTV_result.csv',
                          encoding='UTF-8',
                         index_col='구별')
crime_anal_norm[['인구수','CCTV']] = result_CCTV[['인구수','소계']]
crime_anal_norm.head()
```

```python
col = ['강간','강도','살인','절도','폭력']
crime_anal_norm['범죄'] = np.sum(crime_anal_norm[col], axis=1)
crime_anal_norm.head()
# 1점 만점 범죄 라는 columns을 속에서
```
